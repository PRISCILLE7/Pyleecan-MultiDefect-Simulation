La comparaison entre la machine Toyota Prius de référence et ses variantes modifiées (avec Ntcoil = 80 et défauts) révèle un impact clair du paramètre d’enroulement sur le comportement électromagnétique.
Les courbes de densité de flux magnétique montrent des amplitudes plus élevées et des formes plus perturbées pour les machines modifiées, preuve que les matrices d’enroulement diffèrent bien.
L’analyse fréquentielle met en évidence une augmentation marquée des harmoniques dans les cas modifiés.
Le couple électromagnétique des machines avec Ntcoil = 80 ou défauts est significativement plus élevé (plus négatif), confirmant l’influence directe du nombre de spires.
Les formes d’onde temporelles conservent une cohérence générale mais amplifient les effets liés à la modification de l’enroulement.
