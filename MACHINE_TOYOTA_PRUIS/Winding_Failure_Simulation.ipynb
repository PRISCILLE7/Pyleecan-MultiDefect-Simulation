# Simulating stator winding failures

## Load Machine

```python
%matplotlib inline

# Load the machine
from os.path import join
from pyleecan.Functions.load import load
from pyleecan.definitions import DATA_DIR
from util.simulation import *
from util.failures import *
from IPython.display import display, HTML
```

```python
machine_names=[
   "AUDI_eTron", "BMW_i3", "Railway_Traction", "Renault_Zoe", "Toyota_Prius", "SPMSM_LamSlotMultiWind"
]
for machine_name in machine_names:
    display(HTML(f"<h3>{machine_name}</h3>"))
    machine = load_machine(machine_name)
    simu = load_simulation(name=machine_name+" simulation", machine=machine, rotor_speed=3000)
    out = run_simulation(simu)
    machine.stator.winding.wind_mat[0][0][0][0] = 1
    simu_failure = load_simulation(name=machine_name+" failure simulation", machine=machine, rotor_speed=3000)
    out_failure = run_simulation(simu_failure)
    compare_simulation_results(out1=out, out2=out_failure)
```
