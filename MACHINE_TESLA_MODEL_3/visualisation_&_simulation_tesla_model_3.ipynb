#Visualisation de machine avec defaut de demagnetisation

Avec ce defaut, les aimants n’arrivent plus à fournir le champ magnétique qu’ils donnaient au départ  c.a.d la machine perd en performance.

Conséquences :
    - Baisse du couple produit -> le moteur devient moins puissant.
    - Déséquilibre magnétique -> vibrations ou bruit.
    - Surconsommation -> il faut plus de courant pour obtenir le même couple.
    - Surchauffe -> car la machine travaille “plus fort” pour compenser.

nous avons ici generer des machines avec defaut de demagnetisation partant de 2% to 60/70%

```python
%matplotlib inline

# 1) Chargement 
print("=" * 80)
print("COMPARAISON DÉMAGNÉTISATION - MACHINE RÉFÉRENCE vs DÉMAGNÉTISATION 2%")
print("=" * 80)

# Machine de référence
machine_ref = load(join(DATA_DIR, "Machine", "Tesla_model_3.json"))

# Machine avec défaut de démagnétisation
machine_Demagnetisation1 = load("Tesla_Model3_Dataset_Cohérent_100/Variantes_Defauts/Tesla_Model3_Defaut_Demagnetisation_2_Pourcent.json")

print(" Machines chargées avec succès")

# 2) Analyse comparative des propriétés magnétiques
print("\n ANALYSE COMPARATIVE DES PROPRIÉTÉS MAGNÉTIQUES")
print("=" * 60)

def compare_magnet_properties(machine1, machine2, name1, name2):
    """Compare les propriétés magnétiques de deux machines"""
    
    print(f"\n COMPARAISON: {name1} vs {name2}")
    print("-" * 70)
    
    # === STATOR (inchangé) ===
    print("\n  STATOR:")
    print(f"   Rayon interne: {machine1.stator.Rint*1000:.2f} mm -> {machine2.stator.Rint*1000:.2f} mm")
    print(f"   Rayon externe: {machine1.stator.Rext*1000:.2f} mm -> {machine2.stator.Rext*1000:.2f} mm")
    print(f"   Longueur: {machine1.stator.L1*1000:.2f} mm -> {machine2.stator.L1*1000:.2f} mm")
    
    # === ROTOR (inchangé) ===
    print(f"\n  ROTOR:")
    print(f"   Rayon interne: {machine1.rotor.Rint*1000:.2f} mm -> {machine2.rotor.Rint*1000:.2f} mm")
    print(f"   Rayon externe: {machine1.rotor.Rext*1000:.2f} mm -> {machine2.rotor.Rext*1000:.2f} mm")
    print(f"   Longueur: {machine1.rotor.L1*1000:.2f} mm -> {machine2.rotor.L1*1000:.2f} mm")
    
    # Entrefer
    airgap1 = machine1.stator.Rint - machine1.rotor.Rext
    airgap2 = machine2.stator.Rint - machine2.rotor.Rext
    airgap_variation = (airgap2 - airgap1) / airgap1 * 100
    print(f"\n   ENTREFER: {airgap1*1000:.3f} mm -> {airgap2*1000:.3f} mm ({airgap_variation:+.2f}%)")
    
    # === AIMANTS - ANALYSE DÉTAILLÉE ===
    print(f"\n  AIMANTS - ANALYSE DÉMAGNÉTISATION:")
    print("=" * 50)
    
    if hasattr(machine1.rotor, 'hole') and hasattr(machine2.rotor, 'hole'):
        print(f" Nombre de trous: {len(machine1.rotor.hole)} -> {len(machine2.rotor.hole)}")
        
        total_br_variation = 0
        total_brm20_variation = 0
        total_hc_variation = 0
        magnet_count = 0
        
        for i, (hole1, hole2) in enumerate(zip(machine1.rotor.hole, machine2.rotor.hole)):
            print(f"\n   TROU {i+1} ({type(hole1).__name__}):")
            
            # Analyse des aimants via magnet_dict
            if hasattr(hole1, 'magnet_dict') and hasattr(hole2, 'magnet_dict'):
                magnets1 = hole1.magnet_dict
                magnets2 = hole2.magnet_dict
                
                print(f"     Aimants disponibles: {list(magnets1.keys())}")
                
                for magnet_name in magnets1.keys():
                    if magnet_name in magnets2:
                        magnet1 = magnets1[magnet_name]
                        magnet2 = magnets2[magnet_name]
                        
                        if (magnet1 is not None and magnet2 is not None and 
                            hasattr(magnet1, 'mat_type') and hasattr(magnet2, 'mat_type')):
                            
                            if (hasattr(magnet1.mat_type, 'mag') and 
                                hasattr(magnet2.mat_type, 'mag')):
                                
                                mag1 = magnet1.mat_type.mag
                                mag2 = magnet2.mat_type.mag
                                
                                print(f"\n     {magnet_name.upper()}:")
                                print(f"       Type: {type(magnet1).__name__}")
                                
                                # === PROPRIÉTÉS MAGNÉTIQUES ===
                                
                                # Brm20 (Induction rémanente à 20°C) - PROPRIÉTÉ PRINCIPALE
                                if hasattr(mag1, 'Brm20') and hasattr(mag2, 'Brm20'):
                                    brm20_1 = mag1.Brm20
                                    brm20_2 = mag2.Brm20
                                    brm20_variation = (brm20_2 - brm20_1) / brm20_1 * 100
                                    total_brm20_variation += brm20_variation
                                    magnet_count += 1
                                    
                                    print(f"   Brm20 (Br à 20°C):")
                                    print(f"    Référence: {brm20_1:.4f} T")
                                    print(f"    Démagnétisé: {brm20_2:.4f} T")
                                    print(f"    Variation: {brm20_variation:+.2f}%")
                                    
                                    if abs(brm20_variation + 2.0) < 0.1:
                                        print(f"  Démagnétisation de 2% confirmée")
                                    elif abs(brm20_variation) < 0.01:
                                        print(f" Aucune démagnétisation détectée")
                                    else:
                                        print(f"  Démagnétisation différente de 2% attendu")
                                
                                # Br (Induction rémanente)
                                if hasattr(mag1, 'Br') and hasattr(mag2, 'Br'):
                                    br1 = mag1.Br
                                    br2 = mag2.Br
                                    br_variation = (br2 - br1) / br1 * 100
                                    total_br_variation += br_variation
                                    
                                    print(f"   Br (Induction rémanente):")
                                    print(f"   Référence: {br1:.4f} T")
                                    print(f"    Démagnétisé: {br2:.4f} T")
                                    print(f"    Variation: {br_variation:+.2f}%")
                                
                                # Hc (Champ coercitif)
                                if hasattr(mag1, 'Hc') and hasattr(mag2, 'Hc'):
                                    hc1 = mag1.Hc
                                    hc2 = mag2.Hc
                                    hc_variation = (hc2 - hc1) / hc1 * 100 if hc1 != 0 else 0
                                    total_hc_variation += hc_variation
                                    
                                    print(f"   Hc (Champ coercitif):")
                                    print(f"    Référence: {hc1:.0f} A/m")
                                    print(f"    Démagnétisé: {hc2:.0f} A/m")
                                    print(f"    Variation: {hc_variation:+.2f}%")
                                
                                # mur_lin (Perméabilité relative linéaire)
                                if hasattr(mag1, 'mur_lin') and hasattr(mag2, 'mur_lin'):
                                    mur1 = mag1.mur_lin
                                    mur2 = mag2.mur_lin
                                    mur_variation = (mur2 - mur1) / mur1 * 100 if mur1 != 0 else 0
                                    
                                    print(f"   μr_lin (Perméabilité relative):")
                                    print(f"   Référence: {mur1:.3f}")
                                    print(f"   Démagnétisé: {mur2:.3f}")
                                    print(f"   Variation: {mur_variation:+.2f}%")
                                
                                # === DIMENSIONS DES AIMANTS ===
                                print(f"   DIMENSIONS:")
                                
                                if hasattr(magnet1, 'Hmag') and hasattr(magnet2, 'Hmag'):
                                    hmag1 = magnet1.Hmag * 1000  # en mm
                                    hmag2 = magnet2.Hmag * 1000  # en mm
                                    hmag_variation = (hmag2 - hmag1) / hmag1 * 100
                                    print(f"  Épaisseur: {hmag1:.3f} mm -> {hmag2:.3f} mm ({hmag_variation:+.2f}%)")
                                
                                if hasattr(magnet1, 'Wmag') and hasattr(magnet2, 'Wmag'):
                                    wmag1 = magnet1.Wmag * 1000  # en mm
                                    wmag2 = magnet2.Wmag * 1000  # en mm
                                    wmag_variation = (wmag2 - wmag1) / wmag1 * 100
                                    print(f"   Largeur: {wmag1:.3f} mm -> {wmag2:.3f} mm ({wmag_variation:+.2f}%)")
                                
                                if hasattr(magnet1, 'Lmag') and hasattr(magnet2, 'Lmag'):
                                    lmag1 = magnet1.Lmag * 1000  # en mm
                                    lmag2 = magnet2.Lmag * 1000  # en mm
                                    lmag_variation = (lmag2 - lmag1) / lmag1 * 100
                                    print(f"  Longueur: {lmag1:.3f} mm -> {lmag2:.3f} mm ({lmag_variation:+.2f}%)")
                            else:
                                print(f"  Pas de propriétés magnétiques trouvées")
                        else:
                            print(f"  Aimant non trouvé ou structure différente")
                    else:
                        print(f" {magnet_name} non trouvé dans la machine de référence")
        
        # Résumé de la démagnétisation
        if magnet_count > 0:
            avg_brm20_variation = total_brm20_variation / magnet_count
            avg_br_variation = total_br_variation / magnet_count
            avg_hc_variation = total_hc_variation / magnet_count
            
            print(f"\n   RÉSUMÉ DÉMAGNÉTISATION:")
            print(f"   Nombre d'aimants analysés: {magnet_count}")
            print(f"   Variation moyenne Brm20: {avg_brm20_variation:+.2f}%")
            print(f"   Variation moyenne Br: {avg_br_variation:+.2f}%")
            print(f"   Variation moyenne Hc: {avg_hc_variation:+.2f}%")
            
            if abs(avg_brm20_variation + 2.0) < 0.1:
                print(f"   Démagnétisation de 2% confirmée globalement")
            elif abs(avg_brm20_variation) < 0.01:
                print(f"   Aucune démagnétisation détectée")
            else:
                print(f"  Démagnétisation différente de 2% attendu")
        else:
            print(f"  Aucun aimant trouvé pour l'analyse")
    else:
        print(f"  Pas de trous trouvés dans le rotor")

# Comparaison des propriétés magnétiques
compare_magnet_properties(machine_ref, machine_Demagnetisation1, "Référence", "Démagnétisation 2%")

# 3) Vérification alternative - analyse directe des fichiers JSON
print(f"\n VÉRIFICATION ALTERNATIVE - ANALYSE FICHIERS JSON")
print("=" * 60)

import json

def analyze_json_magnets_detailed(file_path, name):
    """Analyse détaillée des propriétés magnétiques dans un fichier JSON"""
    print(f"\n ANALYSE DÉTAILLÉE {name}:")
    print(f" Fichier: {file_path}")
    
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
        
        if 'rotor' in data and 'hole' in data['rotor']:
            holes = data['rotor']['hole']
            print(f" Nombre de trous: {len(holes)}")
            
            for i, hole in enumerate(holes):
                print(f"\n TROU {i+1}:")
                if 'magnet_dict' in hole:
                    magnets = hole['magnet_dict']
                    print(f"   Aimants trouvés: {list(magnets.keys())}")
                    
                    for magnet_name, magnet_data in magnets.items():
                        print(f"\n   {magnet_name.upper()}:")
                        if 'mat_type' in magnet_data and 'mag' in magnet_data['mat_type']:
                            mag_data = magnet_data['mat_type']['mag']
                            print(f"     Propriétés magnétiques disponibles: {list(mag_data.keys())}")
                            
                            # valeurs importantes
                            for prop in ['Brm20', 'Br', 'Hc', 'mur_lin']:
                                if prop in mag_data:
                                    print(f"     {prop}: {mag_data[prop]}")
                        else:
                            print(f"     Pas de propriétés magnétiques trouvées")
                else:
                    print(f"   Pas de magnet_dict trouvé")
        else:
            print(f" Structure de rotor non trouvée")
            
    except Exception as e:
        print(f" Erreur lors de l'analyse: {e}")

# analyse des deux fichiers en détail
analyze_json_magnets_detailed(join(DATA_DIR, "Machine", "Tesla_model_3.json"), "RÉFÉRENCE")
analyze_json_magnets_detailed("Tesla_Model3_Dataset_Cohérent_100/Variantes_Defauts/Tesla_Model3_Defaut_Demagnetisation_2_Pourcent.json", "DÉMAGNÉTISATION 2%")

# 4) Conclusion
print(f"\n CONCLUSION")
print("=" * 20)

print(" RÉSULTATS DE LA COMPARAISON:")
print("   - Vérification des propriétés magnétiques Brm20, Br, Hc")
print("   - Comparaison des dimensions des aimants")
print("   - Analyse de la cohérence des variations")
print("   - Détection des différences entre les machines")

print("\n" + "=" * 80)
print(" COMPARAISON DÉMAGNÉTISATION TERMINÉE")
print("=" * 80)
```

Defaut materiau:

dans une machine électrique cela  correspond au cas où le matériau utilisé (aimants, tôles du stator ou du rotor) n’a pas les propriétés magnétiques prévues. c.a.d en d'autres terme que la machine n’a pas le comportement électromagnétique attendu, car son matériau est de moins bonne qualité ou a été altéré.

Conséquences:
     - Baisse du couple disponible (car champ magnétique moins fort(Br, HC, Ur)).
     - Consommation plus élevée (il faut plus de courant pour compenser).
     - Echauffement anormal (pertes fer accrues).
     - Dans le cas d’un défaut localisé -> déséquilibre magnétique et vibrations.

Defaut Usinage:
    
ceci correspond à une erreur de fabrication mécanique lors de la production du stator, du rotor, des encoches ou même des aimants.
Concrètement, les dimensions (largeur, profondeur, angle, épaisseur…) ne correspondent pas exactement aux plans.

Conséquences:
 - Variation des performances : couple plus faible ou bruit électromagnétique accru.
 - Harmoniques supplémentaires dans le flux et le courant.
 - Vibrations dues à l’asymétrie.
 - Dans les cas graves : frottement rotor/stator.
