```python
%matplotlib inline

# Load the machine
from os.path import join
from pyleecan.Functions.load import load
from pyleecan.definitions import DATA_DIR
from util.simulation import *
from util.failures import *
from IPython.display import display, HTML
```

```python
# Chargement des machines
machine_ref = load(join(DATA_DIR, "Machine", "Tesla_model_3.json"))
machine_demagnet1 = load("Tesla_Model3_Demagnetisation_Severe/Tesla_Model3_Defaut_Demagnetisation_Severe_16_Pourcent.json")
```

```python
simu_ref = load_simulation(name="ref", machine=machine_ref, rotor_speed=3000, stop=1, num_steps=4000)

simu_demagnet1 = load_simulation(name="demagnet1", machine=machine_demagnet1, rotor_speed=3000, stop=1, num_steps=4000)
```

```python
%matplotlib inline

out_ref = simu_ref.run()
out_demagnet1 = simu_demagnet1.run()

compare_simulation_results(
    out1=out_ref,
    out2=out_demagnet1,
    legend_list=["Tesla_model_3_ref", "demagnet_16_pourcent"]
)
```

```python
out_demagnet1.save(save_path='Tesla_Model3_Demagnetisation_Severe/simulation/demanget_16_pourcent', is_folder=True, type_handle_old=2, type_compression=0)
```
